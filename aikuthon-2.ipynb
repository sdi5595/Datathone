# Finetuning

```python
# let's define model parameters specific to T5
model_params = {
    "MODEL": 'ainize/kobart-news',  # model_type: t5-base/t5-large
    "TRAIN_BATCH_SIZE": 16,  # training batch size
    "VALID_BATCH_SIZE": 16,  # validation batch size
    "EPOCHS": 3,  # number of training epochs
    "LEARNING_RATE": 1e-4,  # learning rate
    "MAX_SOURCE_TEXT_LENGTH": 512,  # max length of source text
    "MAX_TARGET_TEXT_LENGTH": 128,  # max length of target text
    "SEED": 42,  # set seed for reproducibility
}
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from torch.utils.data import DataLoader, Dataset

class TrainDataset(Dataset):
    def __init__(self, df, tokenizer):
        self.document = df["document"]
        self.summary = df["summary"]
        self.tokenizer = tokenizer
        
#         self.max_src_len = self.document.str.len().max()
#         self.max_trg_len = self.summary.str.len().max()
        self.max_src_len = 128
        self.max_trg_len = 64
        
    def __len__(self):
        return len(self.document)
    
    def __getitem__(self, idx):
        src_text = self.document.iloc[idx]
        trg_text = self.summary.iloc[idx]
        
        src = self.tokenizer(
            [src_text],max_length=self.max_src_len, pad_to_max_length=True, truncation=True,
            padding="max_length", return_tensors="pt",)
        trg = self.tokenizer(
            [trg_text],max_length=self.max_trg_len, pad_to_max_length=True, truncation=True,
            padding="max_length", return_tensors="pt",)
        
        src_ids, src_mask =  src["input_ids"].squeeze(), src["attention_mask"].squeeze()
        trg_ids, trg_mask = trg["input_ids"].squeeze(), trg["attention_mask"].squeeze()

        return {
            "source_ids": src_ids,
            "source_mask": src_mask,
            "target_ids": trg_ids,
            "target_mask": trg_mask,
        }
    
    
```

```python
from tqdm import tqdm

def train(epoch, tokenizer, model, device, loader, optimizer):

    """
    Function to be called for training with the parameters passed from main function

    """
    console.clear()
    model.train()
    for _, data in tqdm(enumerate(loader), total=len(loader)):
        y = data["target_ids"].to(device, dtype=torch.long)
        y_ids = y[:, :-1].contiguous()
        lm_labels = y[:, 1:].clone().detach()
        lm_labels[y[:, 1:] == tokenizer.pad_token_id] = -100
        ids = data["source_ids"].to(device, dtype=torch.long)
        mask = data["source_mask"].to(device, dtype=torch.long)

        outputs = model(
            input_ids=ids,
            attention_mask=mask,
            decoder_input_ids=y_ids,
            labels=lm_labels,
        )
        loss = outputs[0]

        if _ % 100 == 0:
            training_logger.add_row(str(epoch), str(_), str(loss))
            print(_)
            console.print(training_logger)

        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        gc.collect()
```

```python
import torch
import numpy as np
import pandas as pd
import torch.nn as nn

    
from transformers import AutoTokenizer, AutoModel

fe_tokenizer = AutoTokenizer.from_pretrained("monologg/kobert")
fe_model = AutoModel.from_pretrained("monologg/kobert")


def validate(epoch, tokenizer, model, device, loader):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    model.eval()
    predictions = []
    actuals = []
    cos_sim = []
    with torch.no_grad():
        for _, data in enumerate(loader, 0):
            y = data['target_ids'].to(device, dtype = torch.long)
            ids = data['source_ids'].to(device, dtype = torch.long)
            mask = data['source_mask'].to(device, dtype = torch.long)

            generated_ids = model.generate(
              input_ids = ids,
              attention_mask = mask, 
              max_length=128, 
              num_beams=2,
              repetition_penalty=2.5, 
              length_penalty=1.0, 
              early_stopping=True
              )
            
# 
            preds = tokenizer.batch_decode(generated_ids, skip_special_tokens=True, clean_up_tokenization_spaces=True)
            targets = tokenizer.batch_decode(y, skip_special_tokens=True, clean_up_tokenization_spaces=True)
            
            if _%100==0:
                console.print(f'Completed {_}')
                print(f"preds : {preds[0]}\ntargets : {targets[0]}")
                
            
            predictions.extend(preds)
            actuals.extend(targets)

    pred_tokens = fe_tokenizer(predictions, return_tensors="pt", padding=True, truncation=True).input_ids
    pred_output = fe_model(pred_tokens).pooler_output.squeeze(0)

    target_tokens = fe_tokenizer(actuals, return_tensors="pt", padding=True, truncation=True).input_ids
    target_output = fe_model(target_tokens).pooler_output.squeeze(0)

    cos = nn.CosineSimilarity(dim=1)
    cos_sim = cos(pred_output, target_output)
            
    return cos_sim
```

```python
# logging
training_logger = Table(
    Column("Epoch", justify="center"),
    Column("Steps", justify="center"),
    Column("Loss", justify="center"),
    title="Training Status",
    pad_edge=False,
    box=box.ASCII,
)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
console.log(f"[Data]: Reading data...\n")

# Importing the raw dataset
# Creation of Dataset and Dataloader
# Defining the train size. So 80% of the data will be used for training and the rest for validation.
from sklearn.model_selection import train_test_split

tmp_data = train_json[['document', 'summary']]
tmp_data['document'] = tmp_data['document'].apply(lambda x:"summarize : " + x)

train_data, val_data = train_test_split(tmp_data, test_size=0.2, random_state=model_params["SEED"])

console.print(f"TRAIN Dataset: {train_data.shape}")
console.print(f"TEST Dataset: {val_data.shape}\n")

# Creating the Training and Validation dataset for further creation of Dataloader
training_set = TrainDataset(train_data, tokenizer)
val_set = TrainDataset(val_data, tokenizer)

# Defining the parameters for creation of dataloaders
train_params = {
    "batch_size": model_params["TRAIN_BATCH_SIZE"],
    "shuffle": True,
    "num_workers": 0,
}

val_params = {
    "batch_size": model_params["VALID_BATCH_SIZE"],
    "shuffle": False,
    "num_workers": 0,
}

# Creation of Dataloaders for testing and validation. This will be used down for training and validation stage for the model.
train_dataloader = DataLoader(training_set, **train_params)
val_dataloader = DataLoader(val_set, **val_params)

# Defining the optimizer that will be used to tune the weights of the network in the training session.
optimizer = torch.optim.AdamW(
    params=model.parameters(), lr=model_params["LEARNING_RATE"]
)

# Training loop
console.log(f"[Initiating Fine Tuning]...\n")

for epoch in range(model_params["EPOCHS"]):
    train(epoch, tokenizer, model, device, train_dataloader, optimizer)
    cos_sim = validate(epoch, tokenizer, model, device, val_dataloader)
    print(sum(cos_sim) / len(cos_sim))

console.log(f"[Saving Model]...\n")
# Saving the model after training
path = os.path.join(output_dir, "model_files")
model.save_pretrained(path)
tokenizer.save_pretrained(path)

# evaluating test dataset
console.log(f"[Initiating Validation]...\n")

    
console.save_text(os.path.join(output_dir, "logs.txt"))

console.log(f"[Validation Completed.]\n")
console.print(
    f"""[Model] Model saved @ {os.path.join(output_dir, "model_files")}\n"""
)
console.print(
    f"""[Validation] Generation on Validation data saved @ {os.path.join(output_dir,'predictions.csv')}\n"""
)
console.print(f"""[Logs] Logs saved @ {os.path.join(output_dir,'logs.txt')}\n""")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from torch.utils.data import DataLoader, Dataset

class TuningDataset(Dataset):
    def __init__(self, df, tokenizer):
        self.document = df["document"]
        self.summary = df["summary"]
        self.tokenizer = tokenizer
        
#         self.max_src_len = self.document.str.len().max()
#         self.max_trg_len = self.summary.str.len().max()
        self.max_src_len = 512
        self.max_trg_len = 128
        
    def __len__(self):
        return len(self.document)//100
    
    def __getitem__(self, idx):
        src_text = self.document.iloc[idx]
        trg_text = self.summary.iloc[idx]
        
        src = self.tokenizer(
            [src_text],max_length=self.max_src_len, pad_to_max_length=True, truncation=True,
            padding="max_length", return_tensors="pt",)
        trg = self.tokenizer(
            [trg_text],max_length=self.max_trg_len, pad_to_max_length=True, truncation=True,
            padding="max_length", return_tensors="pt",)
        
        src_ids, src_mask =  src["input_ids"].squeeze(), src["attention_mask"].squeeze()
        trg_ids, trg_mask = trg["input_ids"].squeeze(), trg["attention_mask"].squeeze()

        return {
            "source_ids": src_ids,
            "source_mask": src_mask,
            "target_ids": trg_ids,
            "target_mask": trg_mask,
        }
    
    
```

```python
def validate(tokenizer, model, device, loader, max_length,min_length, num_beams, length_penalty):
    model.eval()
    predictions = []
    actuals = []
    cos_sim = []
    with torch.no_grad():
        len_ = len(loader)
        for _, data in tqdm(enumerate(loader), total=len_):
            y = data['target_ids'].to(device, dtype = torch.long)
            ids = data['source_ids'].to(device, dtype = torch.long)
            mask = data['source_mask'].to(device, dtype = torch.long)
            
            text_gen_kwargs={"length_penalty":length_penalty,
            "max_length":max_length,
            "num_beams": num_beams,
            "repetition_penalty": 2.5,
            "early_stopping":False}   
                
            generated_ids = model.generate(
              input_ids = ids,
              attention_mask = mask, 
                **text_gen_kwargs
              )
            
            preds = tokenizer.batch_decode(generated_ids, skip_special_tokens=True, clean_up_tokenization_spaces=True)
            targets = tokenizer.batch_decode(y, skip_special_tokens=True, clean_up_tokenization_spaces=True)

            predictions.extend(preds)
            actuals.extend(targets)

    return predictions, actuals
```

```python
torch.manual_seed(0)  # pytorch random seed
np.random.seed(0)  # numpy random seed
torch.backends.cudnn.deterministic = True

from sklearn.model_selection import RandomizedSearchCV

from transformers import AutoTokenizer, AutoModel
import torch.nn as nn

fe_tokenizer = AutoTokenizer.from_pretrained("monologg/kobert")
fe_model = AutoModel.from_pretrained("monologg/kobert")

tuning_params = {
    "batch_size": 32,
    "shuffle": True,
    "num_workers": 0,
}

tuning_set = TuningDataset(train_json, tokenizer)

hyperparams = {
    "max_length": [256],
    "min_length": [0],
    "num_beams": [3, 4],
    "length_penalty": [2.0, 1.0],
}

for num_beams in hyperparams['num_beams']:
    for min_length in hyperparams['min_length']:
        for max_length in hyperparams['max_length']:
            for length_penalty in hyperparams['length_penalty']:
                tuning_loader =  DataLoader(tuning_set, **tuning_params)
                print(f"max_length : {max_length}|min_length : {min_length}|num_beams : {num_beams}|penalty : {length_penalty}|", end="")
                predictions, actuals = validate(tokenizer, model, device,tuning_loader, max_length, min_length, num_beams, length_penalty)

                pooler_prediction_outputs = []
                pooler_actual_outputs = []

                for pred, actual in tqdm(zip(predictions, actuals)):
                    actual_input_ids = fe_tokenizer(actual, return_tensors="pt").input_ids
                    pred_input_ids = fe_tokenizer(pred, return_tensors="pt").input_ids
                    with torch.no_grad():
                        pooler_actual_output = fe_model(actual_input_ids).pooler_output.squeeze(0)
                        pooler_prediction_output = fe_model(pred_input_ids).pooler_output.squeeze(0)
                        pooler_actual_outputs.append(pooler_actual_output) 
                        pooler_prediction_outputs.append(pooler_prediction_output)

                cos = nn.CosineSimilarity(dim=1)
                cos_sim = cos(torch.stack(pooler_actual_outputs).to(device), torch.stack(pooler_prediction_outputs).to(device))

                score = sum(cos_sim) / len(cos_sim)
                print(f"score : {score}")
                
```
